```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, BatchNormalization, Activation
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.constraints import maxnorm
from keras.utils import np_utils
from keras.preprocessing.image import ImageDataGenerator as idg
from skimage import exposure, color
import cv2
from scipy import misc, ndimage

def detail(x,val):
    try:
        print(type(x))
        print(x.dtype)
        print(x.shape)
    except:
        print(type(x))
    if val:
        print(x)
```

```python
# Reading in the data and setting it up
images=np.load('train_and_test.npz')
data=[images[key] for key in images]

# Splitting up the original data into Trainx, Trainy, and test
originalX,originalY,originalTest=data[0],data[1],data[2]
originalYCat=np_utils.to_categorical(data[1])
num_class=len(np.unique(originalY))

# Seeing how the data is organized
for i in range(3):
    print('type(data[%i])=%s'%(i,type(data[i])))
    print('data[%i].dtype=%s'%(i,data[i].dtype))
    print('data[%i].shape=%s'%(i,data[i].shape))
    print('\n')
print('there is %i categories'%num_class)
```

```python
# Splitting up the data based on the classification
classIndexList=[]
for i in range(num_class):
    classIndexList.append(np.argwhere(data[1]==i))
print('classIndexList has a length of %i'%len(classIndexList))

# Showing a single image from x and its label with y
for i in range(len(classIndexList)):
    print('label %i'%i)
    print('classIndexList[%i] details:'%i)
    detail(classIndexList[i],False)
    plt.imshow(originalX[classIndexList[i][0][0]])
    plt.show()
```
